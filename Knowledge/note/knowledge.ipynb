# KNOWLEDGE

A knowledge based agent is an agent that reasons and take action based on an internal representation of information

A knowledge base is a sysytem that conatains some sort of information about the world. Think of it as a database that tells us more about the world. With the information contained in the knowledge base, new information can be gotten from inferences made on the existing information.
A knowledge base agent would have the ability to make these inferences using logical reasoning and would take actions based on the results of these inferences.

### How Knowledge Is Represented In a Computer 
Because computers cannot store and process imformation as plain natural language like humans, some new techniques are developed so computers can understand and work with knowledge.

#### Sentences

1. Logical Representation (E.g. Propositional and predicate logic )
2. Frame-based Representation (object representation)
3. Semantic Networks (graph/node representation with edges representing relationship)

### Models
Given a set of $N$ sentences, there are $2^N$ total ways in which these sentences can be assigned truth values  (as each sentences can either be True or False). A models is a system that can represent any of these combinantion of truth value

### Knowledge Base
A knowledge base as used in propositional (KRL) is a set of proposition or logical combination of propositions that the system knows to be true or false.

### Entailment
A proposition ___Q___ is said to entail another proposition ___P___ if for every ___P___ that is true ___Q___ is also true 
$$
P \models Q
$$
What this means is that whenever P is known to be true, it is always the case that the proposition ___Q___ can be inferred as true from ___P___.
An example is;
>Let ___P___ represent the proposition "It is raining."\
>Let ___Q___ represent the proposition "The ground is wet."

In this case, ___P___ (it is raining) entails ___Q___ (the ground is wet), because if it is true that it is raining, we can infer that the ground must also be wet. So $P\models Q$ is true:\
__NOTE__: The entailment relationship does not impose any restriction on ___Q___ for the case ___P___ is false. The ground may be wet or dry if it didn't rain

### Inference
This is the process of deducing some proposition $ \alpha $ from another proposition or a logical combinantion of a set of proposition ___R___, such that $R \models \alpha$ is true.\
An Example:
> Surpose we have a knowledge base K.B and proposition ___P___, ___Q___, ___T___, such that;\
> ___P___ = It is Tuesday\
> ___R___ = It is raining\
> ___T___ = James would go for a run\
> ___K.B___ =
>> (___P___ $\land$ ___R___) → ___T___ = False\
>>___P___ = True\
>> ___R___ = True

From the knowledge base it is known that it is a rainy Tuesday (___P___ and ___Q___ are true) and it is also known that whenever it is a rainy Tuesday then the event of ___T___("James would go for a run") is False. It is therefore reasonable to infer that since it is a rainy Tuesday then James would not go for a run i.e. $ (\lnot T = True) $

### Goal of a Knowledge based AI agent 
Given a set of proposition $\{P, R, ..., \alpha\}$ all represented in a knowledge base ___K.B___, the objective of a knowledge based AI agent is to check if any of the inferences from the knowledge base  entails $ \alpha $
$$ K.B \models \alpha $$

```python
from logic import *

rain = Assert('rain') # it is raining
uniben = Assert('uniben') # israel went to uniben
lagos = Assert('lagos') # israel went to lagos

kb = KnowledgeBase()
kb.add_k(lagos, Not(And(uniben, lagos)), Imply(Not(rain), uniben), Or(uniben, lagos))  # add knowlegde
model_checking(kb, query=uniben)  # algorithm to check if query is true based on known knowlegde 
```
